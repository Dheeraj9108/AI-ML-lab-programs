```python
class Graph :
    
    def __init__(self,h,graph,sn) :
        self.hv = h
        self.graph = graph
        self.sn = sn
        self.parent = {}
        self.status = {}
        self.sg = {}
    
    def applyAoStar(self) :
        self.aoStar(self.sn,False)
        
    def getNeighbours(self,v) :
        return self.graph.get(v,'')
    
    def getStatus(self,v) :
        return self.status.get(v,0)
    
    def gethv(self,v) :
        return self.hv.get(v,0)
    
    def setStatus(self,v,val) :
        self.status[v] = val
        
    def sethv(self,v,val) :
        self.hv[v] = val
    
    def printSolution(self) :
        print(self.sn)
        print(self.sg)
       
    def calcMinCost(self,v) :
        mincost = 0
        mcnl = {}
        mcnl[mincost] = []
        flag = True
        for Tuple in self.getNeighbours(v) :
            cost = 0
            nodelist = []
            for c,wgt in Tuple :
                cost=cost+self.gethv(c)+wgt
                nodelist.append(c)
                if flag == True :
                    mincost = cost
                    mcnl[mincost] = nodelist
                else :
                    if mincost > cost :
                        mincost  = cost
                        mcnl[mincost] = nodelist
        return mincost,mcnl[mincost]

    def aoStar(self,v,bt) :
        print(self.hv)
        print(self.sg)
        print(v)
        
        if self.getStatus(v)>=0 :
            mincost,cnl = self.calcMinCost(v)
            self.sethv(v,mincost)
            self.setStatus(v,len(cnl))
            
            solved = True
            
            for child in cnl :
                self.parent[child] = v
                if self.getStatus(child)!=-1 :
                    solved = solved & False
                    
            if solved == True :
                self.setStatus(v,-1)
                self.sg[v] = cnl
            
            if v != self.sn :
                self.aoStar(self.parent[v],True)
                
            if bt == False :
                for child in cnl :
                    self.setStatus(child,0)
                    self.aoStar(child,False)

h = {'a':1,'b':6,'c':12,'d':10 ,'e':4, 'f':4,'g':5,'h':7}
graph = {'a':[[('b',1),('c',1)],[('d',1)]],'b':[[('g',1),('h',1)]],'d':[[('e',1),('f',1)]]}
g = Graph(h,graph,'a')
g.applyAoStar()
g.printSolution()
```
